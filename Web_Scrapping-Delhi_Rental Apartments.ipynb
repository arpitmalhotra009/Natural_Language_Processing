```python
from bs4 import BeautifulSoup
from requests import get
import pandas as pd
import time
from collections import OrderedDict
import re
```

```python
headers = ({'User-Agent':
            'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36'})
```

```python
makaan_base="https://www.makaan.com/delhi-residential-property/rent-property-in-delhi-city?page="
makaan_sites =[]
for i in range(1,26):
    makaan_sites.append(makaan_base+str(i))
print(makaan_sites)
```

```python
house_dict=OrderedDict()
```

```python
for idx1, site in enumerate(makaan_sites):
    response = get(site, headers=headers)
    print("Loading Page {} of Makaan.com".format(idx1))
    print("Getting the respone from website")
    print(response, type(response))
    try:
        html_soup = BeautifulSoup(response.text, 'html.parser')
        house_containers = html_soup.find_all('li', class_="cardholder")

        for idx2, house in enumerate(house_containers):
            house_dict[(idx1+1)*(idx2+1)] = [house.find_all('a')[1].text, house.find_all('a')[2].text, house.find_all('td')[2].text, 
                               house.find_all('td')[4].text, house.find_all('li')[0].text, house.find_all('td')[6].text]
        print("Sleeping For 5 seconds!!")
        time.sleep(5)
        print("**************************************************************")
        print("waking up!!")
    except:
        print("Site not responding!! Try again after some time")
        break
```

```python
pd.DataFrame(house_dict).head()
```

## Transposing the Data Frame

```python
data = pd.DataFrame(house_dict).T
```

```python
data.head()
```

## Adding the columns

```python
data.columns = ["Description_BHK", "Address", "Rent_in_INR", "Area_in_sqft", "Deposit", "Furnish_type"]
```

```python
data.head()
```

## For Changing String to integer

```python
def dep_change(x):
    if x =='No Deposit':
        z= 0.0
    else:
        z = re.sub("[A-Za-z\s\,]", "", x)
        z= float(z)
        if z<10:
            z=z*100000
        else:
            pass
    return z
```

```python
data['Rent_in_INR']=data['Rent_in_INR'].apply(dep_change)
data['Deposit']=data['Deposit'].apply(dep_change)
```
